```python
import mne
import numpy as np
import os

# List of EEG dataset file paths
vs_file_paths = [
    '/content/VegetativeState_05.set',
    '/content/VegetativeState_06.set',
    '/content/VegetativeState_07.set',
    '/content/VegetativeState_27.set',
    '/content/VegetativeState_44.set'
]

def check_file_integrity(file_path):
    """Check if both .set and .fdt files exist."""
    fdt_path = file_path.replace('.set', '.fdt')
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"EEG dataset (.set) file not found: {file_path}")
    if not os.path.exists(fdt_path):
        print(f"‚ö†Ô∏è Warning: EEG binary data (.fdt) file missing for {file_path}. Attempting to proceed with metadata only.")
    else:
        print(f"‚úÖ {file_path} and {fdt_path} found. Proceeding with loading.")

def get_actual_sample_count(file_path):
    """Estimate actual sample count from .fdt file size."""
    fdt_path = file_path.replace('.set', '.fdt')

    # If .fdt file is missing, return None (indicating we cannot compute actual sample count)
    if not os.path.exists(fdt_path):
        return None

    file_size = os.path.getsize(fdt_path)  # File size in bytes
    dtype_size = 4  # float32 (4 bytes per sample)

    raw = mne.io.read_raw_eeglab(file_path, preload=False, verbose=False)
    n_channels = raw.info['nchan']

    actual_samples = file_size // (n_channels * dtype_size)
    return actual_samples

def load_and_correct_data(file_path):
    """Load EEG data from an EEGLAB .set file, correcting sample mismatches."""
    try:
        # Load metadata only first
        raw = mne.io.read_raw_eeglab(file_path, preload=False, verbose=True)
        print(f"üìÇ File {file_path} metadata loaded successfully.")

        # Expected vs actual samples
        expected_samples = raw.n_times
        actual_samples = get_actual_sample_count(file_path)

        if actual_samples is None:
            print(f"‚ö†Ô∏è Unable to verify actual sample count for {file_path} (missing .fdt file). Proceeding with expected metadata.")
            raw.load_data()
            return raw

        print(f"üîç Expected samples: {expected_samples}, Actual samples: {actual_samples}")

        # Load the actual EEG data safely
        raw.load_data()
        eeg_data = raw.get_data()

        # Handle sample mismatch
        if expected_samples != actual_samples:
            print(f"‚ö†Ô∏è Sample mismatch detected in {file_path}. Attempting correction...")

            if actual_samples < expected_samples:
                # Case 1: Actual data is shorter ‚Üí Pad with zeros
                padding = np.zeros((raw.info['nchan'], expected_samples - actual_samples))
                corrected_data = np.hstack((eeg_data, padding))
                print(f"üõ†Ô∏è Padded missing samples with zeros.")

            else:
                # Case 2: Actual data is longer ‚Üí Truncate extra samples
                corrected_data = eeg_data[:, :expected_samples]
                print(f"‚úÇÔ∏è Truncated excess samples.")

            # Create a new RawArray with the corrected data
            raw_corrected = mne.io.RawArray(corrected_data, raw.info)
            raw = raw_corrected

        return raw

    except Exception as e:
        print(f"‚ùå Error loading file {file_path}: {e}")
        return None

def handle_boundary_events(raw):
    """Handle boundary events to prevent issues in filtering and epoching."""
    annotations = raw.annotations
    if len(annotations) > 0:
        print(f"‚ö†Ô∏è {len(annotations)} boundary events found. Handling discontinuities...")

        new_descriptions = [desc for desc in annotations.description if "boundary" not in desc]
        new_onsets = [onset for i, onset in enumerate(annotations.onset) if "boundary" not in annotations.description[i]]
        new_durations = [dur for i, dur in enumerate(annotations.duration) if "boundary" not in annotations.description[i]]

        raw.set_annotations(mne.Annotations(onset=new_onsets, duration=new_durations, description=new_descriptions))

    return raw

def handle_nan_values(raw):
    """Handle NaN values and bad channels."""
    if np.any(np.isnan(raw.get_data())):
        print("‚ö†Ô∏è NaN values detected. Interpolating bad channels...")
        raw.info['bads'] = ['EEG 001']  # Update with actual bad channels if known
        raw.interpolate_bads()
    return raw

def save_corrected_data(raw, corrected_file_path):
    """Save the corrected EEG data in FIF format."""
    raw.save(corrected_file_path, overwrite=True)
    print(f"‚úÖ Corrected file saved as: {corrected_file_path}")

def process_eeglab_files(file_paths):
    """Process all EEGLAB (.set) files safely."""
    for file_path in file_paths:
        try:
            check_file_integrity(file_path)
            raw = load_and_correct_data(file_path)

            if raw is not None:
                raw = handle_boundary_events(raw)
                raw = handle_nan_values(raw)

                corrected_file_path = file_path.replace('.set', '_corrected.fif')
                save_corrected_data(raw, corrected_file_path)

                raw.plot()  # Optional visualization

        except Exception as e:
            print(f"üö® Error processing {file_path}: {e}")

if __name__ == "__main__":
    process_eeglab_files(vs_file_paths)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mne

def calculate_power_ratios(epochs):
    """Calculate power spectral ratios for each epoch."""
    # Define frequency bands
    bands = {
        'delta': (0.5, 4),
        'theta': (4, 8),
        'alpha': (8, 12),
        'beta': (12, 30),
        'gamma': (30, 45)
    }

    # Compute PSD for all epochs at once
    spectrum = epochs.compute_psd(method='welch', n_fft=512, n_overlap=256)  # Use smaller n_fft (512) and n_overlap (256)

    # Get the PSD data and frequency bins
    psds = spectrum.get_data()  # Shape: (n_epochs, n_channels, n_freqs)
    freqs = spectrum.freqs

    # Calculate power ratios for each band
    power_ratios = {}
    for band_name, (low_freq, high_freq) in bands.items():
        band_mask = (freqs >= low_freq) & (freqs <= high_freq)
        band_power = np.trapz(psds[:, :, band_mask], axis=2)  # Integrate across frequencies
        power_ratios[band_name] = band_power

    # Normalize power ratios to get relative power
    total_power = np.sum([np.sum(band_power, axis=1) for band_power in power_ratios.values()], axis=0)
    normalized_ratios = {band: band_power / total_power[:, None] for band, band_power in power_ratios.items()}

    return normalized_ratios

def binarize_signal(signal, threshold=None):
    """Binarize a signal based on its mean or a specified threshold."""
    if threshold is None:
        threshold = np.mean(signal)
    return (signal > threshold).astype(int)

def lempel_ziv_complexity(binary_signal):
    """Calculate the Lempel-Ziv complexity of a binary signal."""
    n = len(binary_signal)
    i, lz_complexity = 0, 1
    substrings = set()
    current_substring = ""

    while i < n:
        current_substring += str(binary_signal[i])
        if current_substring not in substrings:
            substrings.add(current_substring)
            current_substring = ""
            lz_complexity += 1
        i += 1

    return lz_complexity

def analyze_subject_data(epochs):
    """Analyze EEG data for power ratios and Lempel-Ziv complexity."""
    # Calculate power spectral ratios
    power_ratios = calculate_power_ratios(epochs)

    # Get all epoch data (shape: n_epochs x n_channels x n_times)
    all_epoch_data = epochs.get_data()

    # Calculate Lempel-Ziv complexity for each epoch
    lz_complexities = []
    for epoch_data in all_epoch_data:
        # Use the first channel (or modify as needed, e.g., average across channels)
        signal = epoch_data[0]
        binary_signal = binarize_signal(signal)
        lz_complexity_value = lempel_ziv_complexity(binary_signal)
        lz_complexities.append(lz_complexity_value)

    return power_ratios, lz_complexities

def plot_lzc_values(lz_complexities, patient_id, title="Lempel-Ziv Complexity (LZC)"):
    """Plot Lempel-Ziv Complexity (LZC) values."""
    plt.figure(figsize=(10, 6))
    plt.plot(lz_complexities, marker='o', linestyle='-', color='b')
    plt.title(f"{title} for Patient {patient_id}")
    plt.xlabel("Epoch Index")
    plt.ylabel("LZC Value")
    plt.grid()
    plt.show()

# Simulate data for 2 VS patients
sfreq = 256  # Sampling frequency in Hz
n_channels = 5  # Adjust number of channels based on your data
rng = np.random.RandomState(42)

n_patients = 5

# Specify the number of epochs for each patient
epochs_per_patient = {
    1: 2890,  # VS Patient 1
    2: 2899,   # VS Patient 2
    3: 2873,  # VS Patient 3
    4: 2882,  # VS Patient 4
    4: 2880  # VS Patient 5
}

n_times = int(sfreq * 8)  # 8 seconds of data per epoch

# Create channel names based on the number of channels
ch_names = [f'EEG{i+1}' for i in range(n_channels)]
info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types=['eeg'] * n_channels)

# Store results for both VS patients
all_power_ratios = {}
all_lz_complexities = {}

# Analyze data for both VS patients
for patient_id in range(1, n_patients + 1):
    n_epochs = epochs_per_patient[patient_id]  # Get the number of epochs for the current patient

    # Simulate data for each patient (replace with actual data loading if available)
    data_patient = rng.randn(n_epochs, n_channels, n_times)

    # Create EpochsArray object for each patient
    epochs = mne.EpochsArray(data_patient, info)

    # Analyze data for the current patient
    power_ratios, lz_complexities = analyze_subject_data(epochs)

    # Store results
    all_power_ratios[patient_id] = power_ratios
    all_lz_complexities[patient_id] = lz_complexities

    # Plot LZC values for the current patient
    plot_lzc_values(lz_complexities, patient_id, title="LZC Values for All Epochs")

# Optional: Display or save the power ratios and LZC values for each patient
# For example, print the power ratios for the first patient:
print("Power Ratios for Patient 1:", all_power_ratios[1])
```

```python
import mne
from mne.preprocessing import ICA

def clean_ica_epochs(patient_file, exclude_components, output_filename):
    """Load EEG epochs, apply ICA to remove artifacts, and save cleaned data."""

    # Load preprocessed epochs
    epochs = mne.read_epochs(patient_file, preload=True)

    # Rename channels to match standard 10-20 naming
    rename_dict = {
        "C3:REF": "C3",
        "C4:REF": "C4",
        "Fz:REF": "Fz",
        "Cz:REF": "Cz",
        "Pz:REF": "Pz",
    }
    epochs.rename_channels(rename_dict)

    # Set standard electrode positions using the 10-20 system BEFORE ICA
    montage = mne.channels.make_standard_montage("standard_1020")
    epochs.set_montage(montage)

    # Apply ICA using FastICA, choosing the number of components dynamically
    ica = ICA(n_components=0.99, method="fastica", random_state=42)  # Keep 99% variance
    ica.fit(epochs)

    # Optional: Plot ICA components for manual inspection (Comment this out for batch processing)
    # ica.plot_components()

    # Exclude components suspected to be artifacts based on inspection
    ica.exclude = exclude_components

    # Apply ICA to remove artifacts
    epochs_clean = ica.apply(epochs)

    # Save cleaned epochs
    epochs_clean.save(output_filename, overwrite=True)
    print(f"‚úÖ Cleaned epochs saved as {output_filename}")

# List of patients' data and component exclusions
patients = [
    ("VS_Patient_1-epo.fif", [1, 3, 4], "VS_Patient_1-epo-clean.fif"),
    ("VS_Patient_2-epo.fif", [0, 1, 4], "VS_Patient_2-epo-clean.fif"),
    ("VS_Patient_3-epo.fif", [0, 1, 2, 4], "VS_Patient_3-epo-clean.fif"),
    ("VS_Patient_4-epo.fif", [1, 3, 4], "VS_Patient_4-epo-clean.fif"),
    ("VS_Patient_5-epo.fif", [0, 2, 4], "VS_Patient_5-epo-clean.fif"),
]

# Loop through each patient and clean their epochs
for patient_file, exclude_components, output_filename in patients:
    clean_ica_epochs(patient_file, exclude_components, output_filename)

```

```python
import mne
from sklearn.preprocessing import StandardScaler

# List of the cleaned epoch file names for all 5 patients
patient_files = [
    "VS_Patient_1-epo-clean.fif",
    "VS_Patient_2-epo-clean.fif",
    "VS_Patient_3-epo-clean.fif",
    "VS_Patient_4-epo-clean.fif",
    "VS_Patient_5-epo-clean.fif"
]

# Loop through each patient's dataset
for patient_file in patient_files:
    # Load the cleaned epochs for the current patient
    epochs_clean = mne.read_epochs(patient_file, preload=True)

    # Extract the EEG data as a NumPy array (channels x time x trials)
    data = epochs_clean.get_data()  # Shape: (n_epochs, n_channels, n_times)

    # Reshape the data to 2D: (n_epochs * n_channels) x n_times for normalization
    n_epochs, n_channels, n_times = data.shape
    data_reshaped = data.reshape(-1, n_times)  # Flatten epochs and channels

    # Normalize (standardize) the data: mean = 0, std = 1 for each channel
    scaler = StandardScaler()
    data_normalized = scaler.fit_transform(data_reshaped)

    # Reshape back to the original shape (n_epochs, n_channels, n_times)
    data_normalized = data_normalized.reshape(n_epochs, n_channels, n_times)

    # Set the normalized data back to the epochs
    epochs_clean._data = data_normalized

    # Save the normalized epochs with a new name
    normalized_filename = patient_file.replace("-epo-clean.fif", "-epo-normalized.fif")
    epochs_clean.save(normalized_filename, overwrite=True)

    print(f"Normalized epochs saved for {patient_file} as {normalized_filename}")

```

```python
import os
import numpy as np
import pandas as pd
import mne

# Define frequency bands
freq_bands = {
    "delta": (0.5, 4),
    "theta": (4, 8),
    "alpha": (8, 13),
    "beta": (13, 30),
    "gamma": (30, 45)
}

def compute_band_power(psd, freqs, band):
    """Compute average power in a specific frequency band."""
    fmin, fmax = freq_bands[band]

    # Get valid frequency indices
    band_indices = np.where((freqs >= fmin) & (freqs <= fmax))[0]
    if len(band_indices) == 0:
        print(f"‚ö†Ô∏è No valid frequency bins for {band} band.")
        return np.nan  # Return NaN if no valid frequency bins are available

    # Compute band power across epochs and channels
    band_power = np.mean(psd[:, :, band_indices], axis=(0, 2))  # Mean over epochs & frequency bins
    return band_power

def analyze_psd(eeg_file):
    """Compute power spectral density (PSD) and extract band power."""
    try:
        if not os.path.exists(eeg_file):
            print(f"‚ùå File not found: {eeg_file}")
            return None

        print(f"üìÇ Processing: {eeg_file}")

        # Load EEG epochs
        epochs = mne.read_epochs(eeg_file, preload=True)

        # Select EEG channels
        epochs.pick_types(eeg=True)

        # Compute PSD using Welch‚Äôs method
        psd_data = epochs.compute_psd(method="welch", fmin=0.5, fmax=45, n_fft=1024)
        psd, freqs = psd_data.get_data(return_freqs=True)

        # Ensure correct shape
        if psd.shape[2] != len(freqs):
            print(f"‚ö†Ô∏è Mismatch: PSD shape {psd.shape}, Frequency shape {freqs.shape}")
            return None  # Skip processing if shapes mismatch

        # Compute band power
        band_powers = {band: compute_band_power(psd, freqs, band) for band in freq_bands}
        band_power_mean = {band: np.nanmean(power) for band, power in band_powers.items()}

        return band_power_mean

    except Exception as e:
        print(f"‚ùå Error processing {eeg_file}: {e}")
        return None

def process_all_files(folder_path):
    """Process all EEG files in a given folder."""
    results = []

    for file_name in os.listdir(folder_path):
        if file_name.endswith("-epo.fif"):  # Ensure correct file format
            eeg_file = os.path.join(folder_path, file_name)
            band_power = analyze_psd(eeg_file)
            if band_power:
                band_power["Filename"] = file_name
                results.append(band_power)

    # Convert results to a DataFrame and save
    df = pd.DataFrame(results)
    output_file = os.path.join(folder_path, "vs_psd_results.csv")
    df.to_csv(output_file, index=False)

    print(f"‚úÖ PSD results saved to '{output_file}'.")

# Run processing on your dataset
data_folder = "/content/drive/MyDrive/ADSP_VS"
process_all_files(data_folder)


```

```python
import mne
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import antropy as ant  # Library for entropy calculations
from joblib import Parallel, delayed  # Parallel Processing
from google.colab import drive  # Google Drive support

# üìÇ **Mount Google Drive**
drive.mount('/content/drive')

# üìÇ **Specify 5 EEG .fif Files**
file_paths = [
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_1-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_2-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_3-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_4-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_5-epo-normalized.fif"
]

# üß† **Define EEG Bands (Hz)**
eeg_bands = {
    "delta": (1, 4),
    "theta": (4, 8),
    "alpha": (8, 12),
    "beta": (12, 30),
    "gamma": (30, 45)
}

# üìä **Initialize an Empty List to Store Entropy Values**
entropy_data = []

# üöÄ **Process Each File**
for file in file_paths:
    print(f"üîÑ Loading file: {file}")
    epochs = mne.read_epochs(file, preload=True)

    # üîΩ **Fix Nyquist Frequency Issue by Resampling**
    new_sfreq = max(100, 2 * max([high for _, high in eeg_bands.values()]))  # Ensure at least 90 Hz
    epochs.resample(new_sfreq, npad="auto")  # Resample to fix Nyquist limit

    # üìâ **Use a Subset of Epochs for Faster Computation**
    n_epochs = min(100, len(epochs))  # Use max 100 epochs
    epochs = epochs[:n_epochs]

    raw_data = epochs.get_data()  # Shape: (n_epochs, n_channels, n_times)

    # ‚úÖ **Reduce Dimensionality: Compute Mean Across Channels**
    raw_data = np.mean(raw_data, axis=1)  # Shape: (n_epochs, n_times)

    # **Parallel Function to Compute Entropy**
    def compute_entropy(epoch):
        return {
            "ApEn": ant.app_entropy(epoch),
            "SampEn": ant.sample_entropy(epoch)
        }

    # **Compute Entropy for Each Band**
    for band, (low, high) in eeg_bands.items():
        filtered_data = mne.filter.filter_data(
            raw_data, sfreq=epochs.info['sfreq'], l_freq=low, h_freq=high, verbose=False
        )

        # üî• **Parallel Entropy Computation**
        entropy_results = Parallel(n_jobs=-1)(
            delayed(compute_entropy)(epoch) for epoch in filtered_data
        )

        # Convert list of dicts to separate lists
        ApEn_vals = [res["ApEn"] for res in entropy_results]
        SampEn_vals = [res["SampEn"] for res in entropy_results]

        # Store mean entropy values
        entropy_data.append({
            "Filename": file.split("/")[-1],
            "Band": band,
            "ApEn": np.mean(ApEn_vals),
            "SampEn": np.mean(SampEn_vals)
        })

# üìë **Convert to DataFrame**
df_entropy = pd.DataFrame(entropy_data)

# üîÑ **Convert from Wide to Long Format for Seaborn**
df_entropy_melted = df_entropy.melt(
    id_vars=["Filename", "Band"],
    var_name="Entropy Type",
    value_name="Entropy Value"
)

# üé® **Plot with Improved Readability**
plt.figure(figsize=(12, 6))

sns.barplot(
    x="Entropy Type",
    y="Entropy Value",
    hue="Band",
    data=df_entropy_melted,
    capsize=0.1,
    errcolor="black",
    errwidth=1.5
)

# üñå **Styling for Better Readability**
plt.title("Distribution of Approximate Entropy and Sample Entropy Across EEG Files", fontsize=14, fontweight='bold')
plt.xlabel("Entropy Type", fontsize=12)
plt.ylabel("Entropy Value", fontsize=12)
plt.xticks(rotation=15, fontsize=11)
plt.yticks(fontsize=11)
plt.legend(title="EEG Bands", fontsize=11, title_fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.6)

# üìà **Show the Plot**
plt.show()

# üìÅ **Save Data to Google Drive**
save_path = "/content/drive/MyDrive/ADSP_VS/Entropy_Results.csv"
df_entropy.to_csv(save_path, index=False)
print(f"‚úÖ Entropy results saved to: {save_path}")

```

```python
import mne
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.signal import hilbert  # Hilbert Transform for PLV
from google.colab import drive  # Google Drive support

# üìÇ **Mount Google Drive**
drive.mount('/content/drive')

# üìÇ **Specify 5 EEG .fif Files**
file_paths = [
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_1-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_2-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_3-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_4-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/VS_Patient_5-epo-normalized.fif"
]

# üß† **Define EEG Bands (Hz)**
eeg_bands = {
    "delta": (1, 4),
    "theta": (4, 8),
    "alpha": (8, 12),
    "beta": (12, 30),
    "gamma": (30, 45)
}

# üìä **Initialize List for PLV Storage**
plv_data = []

# üöÄ **Process Each File**
for file in file_paths:
    print(f"üîÑ Loading file: {file}")
    epochs = mne.read_epochs(file, preload=True)

    # üîΩ **Fix Nyquist Frequency Issue by Resampling**
    new_sfreq = max(100, 2 * max([high for _, high in eeg_bands.values()]))
    epochs.resample(new_sfreq, npad="auto")

    # üìâ **Use a Subset of Epochs for Faster Computation**
    n_epochs = min(100, len(epochs))
    epochs = epochs[:n_epochs]

    raw_data = epochs.get_data()  # Shape: (n_epochs, n_channels, n_times)

    # üß† **Compute PLV for Functional Connectivity**
    def compute_plv(data, sfreq, low, high):
        """Compute PLV between all channel pairs for a given frequency band."""
        filtered = mne.filter.filter_data(data, sfreq=sfreq, l_freq=low, h_freq=high, verbose=False)
        analytic_signal = hilbert(filtered)  # Apply Hilbert Transform
        phase_angles = np.angle(analytic_signal)  # Extract phase

        n_channels = phase_angles.shape[0]
        plv_matrix = np.zeros((n_channels, n_channels))  # PLV storage

        # Compute PLV for each channel pair
        for i in range(n_channels):
            for j in range(i + 1, n_channels):
                phase_diff = phase_angles[i] - phase_angles[j]
                plv = np.abs(np.mean(np.exp(1j * phase_diff)))  # Compute PLV
                plv_matrix[i, j] = plv
                plv_matrix[j, i] = plv  # Symmetric matrix

        return np.mean(plv_matrix[np.triu_indices(n_channels, k=1)])  # Mean PLV

    # **Compute PLV for Each EEG Band**
    for band, (low, high) in eeg_bands.items():
        plv_value = compute_plv(raw_data, epochs.info['sfreq'], low, high)

        plv_data.append({
            "Filename": file.split("/")[-1],
            "Band": band,
            "PLV": plv_value
        })

# üìë **Convert to DataFrame**
df_plv = pd.DataFrame(plv_data)

# üé® **Plot PLV Values**
plt.figure(figsize=(12, 6))
sns.barplot(
    x="Band",
    y="PLV",
    data=df_plv,
    capsize=0.1,
    errcolor="black",
    errwidth=1.5
)
plt.title("Phase Locking Value (PLV) Across EEG Bands", fontsize=14, fontweight='bold')
plt.xlabel("EEG Band", fontsize=12)
plt.ylabel("Mean PLV", fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.show()

# üìÅ **Save Data to Google Drive**
save_path = "/content/drive/MyDrive/ADSP_VS/PLV_Results.csv"
df_plv.to_csv(save_path, index=False)
print(f"‚úÖ PLV results saved to: {save_path}")

```

```python
import mne
from mne.preprocessing import ICA

def clean_ica_epochs(subject_file, exclude_components, output_filename):
    """Load EEG epochs, apply ICA to remove artifacts, and save cleaned data."""

    # Load preprocessed epochs
    epochs = mne.read_epochs(subject_file, preload=True)

    # Print actual channel names
    print(f"\nüìå Available channels in {subject_file}: {epochs.ch_names}")

    # Define a renaming dictionary for EEG channels
    rename_dict = {
        "EEG Fpz-Cz": "Fpz-Cz",
        "EEG Pz-Oz": "Pz-Oz"
    }

    # Rename only existing EEG channels
    rename_dict = {old: new for old, new in rename_dict.items() if old in epochs.ch_names}
    if rename_dict:
        epochs.rename_channels(rename_dict)
        print(f"‚úÖ Renamed channels: {rename_dict}")

    # Set correct channel types for non-EEG channels
    non_eeg_types = {
        "EOG horizontal": "eog",
        "Resp oro-nasal": "resp",
        "EMG submental": "emg",
        "Temp rectal": "misc",
        "Event marker": "stim"
    }

    # Filter only existing channels and update types
    non_eeg_types = {ch: ch_type for ch, ch_type in non_eeg_types.items() if ch in epochs.ch_names}
    epochs.set_channel_types(non_eeg_types)
    print(f"‚úÖ Set channel types: {non_eeg_types}")

    # Drop non-EEG channels before setting the montage
    drop_channels = list(non_eeg_types.keys())
    if drop_channels:
        epochs.drop_channels(drop_channels)
        print(f"‚ùå Dropped non-EEG channels: {drop_channels}")

    # Set montage (only for EEG channels)
    montage = mne.channels.make_standard_montage("standard_1020")
    epochs.set_montage(montage, on_missing="ignore")

    # Apply ICA (Keep 99% variance)
    ica = ICA(n_components=0.99, method="fastica", random_state=42)
    ica.fit(epochs)

    # Exclude artifact components
    ica.exclude = exclude_components

    # Apply ICA to remove artifacts
    epochs_clean = ica.apply(epochs)

    # Save cleaned epochs
    epochs_clean.save(output_filename, overwrite=True)
    print(f"‚úÖ Cleaned epochs saved as {output_filename}")

# List of healthy subjects' data and component exclusions
healthy_subjects = [
    ("Healthy_Subject_1-epo.fif", [1, 3, 4], "Healthy_Subject_1-epo-clean.fif"),
    ("Healthy_Subject_2-epo.fif", [0, 1, 4], "Healthy_Subject_2-epo-clean.fif"),
    ("Healthy_Subject_3-epo.fif", [0, 1, 2, 4], "Healthy_Subject_3-epo-clean.fif"),
    ("Healthy_Subject_4-epo.fif", [1, 3, 4], "Healthy_Subject_4-epo-clean.fif"),
    ("Healthy_Subject_5-epo.fif", [0, 2, 4], "Healthy_Subject_5-epo-clean.fif"),
    ("Healthy_Subject_6-epo.fif", [1, 2, 3], "Healthy_Subject_6-epo-clean.fif"),
    ("Healthy_Subject_7-epo.fif", [0, 1, 3], "Healthy_Subject_7-epo-clean.fif"),
    ("Healthy_Subject_8-epo.fif", [2, 4, 5], "Healthy_Subject_8-epo-clean.fif"),
]

# Loop through each healthy subject and clean their epochs
for subject_file, exclude_components, output_filename in healthy_subjects:
    clean_ica_epochs(subject_file, exclude_components, output_filename)
```

```python
import mne
from sklearn.preprocessing import StandardScaler

# List of the cleaned epoch file names for all 8 healthy subjects
healthy_subject_files = [
    "Healthy_Subject_1-epo-clean.fif",
    "Healthy_Subject_2-epo-clean.fif",
    "Healthy_Subject_3-epo-clean.fif",
    "Healthy_Subject_4-epo-clean.fif",
    "Healthy_Subject_5-epo-clean.fif",
    "Healthy_Subject_6-epo-clean.fif",
    "Healthy_Subject_7-epo-clean.fif",
    "Healthy_Subject_8-epo-clean.fif"
]

# Loop through each healthy subject's dataset
for subject_file in healthy_subject_files:
    # Load the cleaned epochs for the current subject
    epochs_clean = mne.read_epochs(subject_file, preload=True)

    # Extract EEG data as a NumPy array (n_epochs, n_channels, n_times)
    data = epochs_clean.get_data()

    # Reshape data for normalization: (n_epochs * n_channels) x n_times
    n_epochs, n_channels, n_times = data.shape
    data_reshaped = data.reshape(-1, n_times)

    # Normalize (standardize) the data: mean = 0, std = 1 for each time point
    scaler = StandardScaler()
    data_normalized = scaler.fit_transform(data_reshaped)

    # Reshape back to original shape (n_epochs, n_channels, n_times)
    data_normalized = data_normalized.reshape(n_epochs, n_channels, n_times)

    # Set the normalized data back to the epochs object
    epochs_clean._data = data_normalized

    # Save the normalized epochs with a new name
    normalized_filename = subject_file.replace("-epo-clean.fif", "-epo-normalized.fif")
    epochs_clean.save(normalized_filename, overwrite=True)

    print(f"‚úÖ Normalized epochs saved for {subject_file} as {normalized_filename}")
```

```python
import os
import numpy as np
import pandas as pd
import mne

# Define frequency bands
freq_bands = {
    "delta": (0.5, 4),
    "theta": (4, 8),
    "alpha": (8, 13),
    "beta": (13, 30),
    "gamma": (30, 45)
}

def compute_band_power(psd, freqs, band):
    """Compute average power in a specific frequency band."""
    fmin, fmax = freq_bands[band]

    # Get valid frequency indices
    band_indices = np.where((freqs >= fmin) & (freqs <= fmax))[0]
    if len(band_indices) == 0:
        print(f"‚ö†Ô∏è No valid frequency bins for {band} band.")
        return np.nan  # Return NaN if no valid frequency bins are available

    # Compute band power across epochs and channels
    band_power = np.mean(psd[:, :, band_indices], axis=(0, 2))  # Mean over epochs & frequency bins
    return band_power

def analyze_psd(eeg_file):
    """Compute power spectral density (PSD) and extract band power."""
    try:
        if not os.path.exists(eeg_file):
            print(f"‚ùå File not found: {eeg_file}")
            return None

        print(f"üìÇ Processing: {eeg_file}")

        # Load EEG epochs
        epochs = mne.read_epochs(eeg_file, preload=True)

        # Select EEG channels
        epochs.pick_types(eeg=True)

        # Compute PSD using Welch‚Äôs method
        psd_data = epochs.compute_psd(method="welch", fmin=0.5, fmax=45, n_fft=1024)
        psd, freqs = psd_data.get_data(return_freqs=True)

        # Ensure correct shape
        if psd.shape[2] != len(freqs):
            print(f"‚ö†Ô∏è Mismatch: PSD shape {psd.shape}, Frequency shape {freqs.shape}")
            return None  # Skip processing if shapes mismatch

        # Compute band power
        band_powers = {band: compute_band_power(psd, freqs, band) for band in freq_bands}
        band_power_mean = {band: np.nanmean(power) for band, power in band_powers.items()}

        return band_power_mean

    except Exception as e:
        print(f"‚ùå Error processing {eeg_file}: {e}")
        return None

def process_all_files(folder_path):
    """Process all EEG files in a given folder."""
    results = []

    for file_name in os.listdir(folder_path):
        if file_name.endswith("-epo.fif"):  # Ensure correct file format
            eeg_file = os.path.join(folder_path, file_name)
            band_power = analyze_psd(eeg_file)
            if band_power:
                band_power["Filename"] = file_name
                results.append(band_power)

    # Convert results to a DataFrame and save
    df = pd.DataFrame(results)
    output_file = os.path.join(folder_path, "healthy_psd_results.csv")
    df.to_csv(output_file, index=False)

    print(f"‚úÖ PSD results saved to '{output_file}'.")

# Run processing on your dataset
healthy_data_folder = "/content/drive/MyDrive/ADSP_VS"
process_all_files(healthy_data_folder)
```

```python
import mne
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import antropy as ant  # Library for entropy calculations
from joblib import Parallel, delayed  # Parallel Processing
from google.colab import drive  # Google Drive support

# üìÇ **Mount Google Drive**
drive.mount('/content/drive')

# üìÇ **Specify Healthy EEG .fif Files**
healthy_file_paths = [
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_1-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_2-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_3-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_4-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_5-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_6-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_7-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_8-epo-normalized.fif"
]

# üß† **Define EEG Bands (Hz)**
eeg_bands = {
    "delta": (1, 4),
    "theta": (4, 8),
    "alpha": (8, 12),
    "beta": (12, 30),
    "gamma": (30, 45)
}

# üìä **Initialize an Empty List to Store Entropy Values**
entropy_data = []

# üöÄ **Process Each Healthy Subject File**
for file in healthy_file_paths:
    print(f"üîÑ Loading file: {file}")
    epochs = mne.read_epochs(file, preload=True)

    # üîΩ **Fix Nyquist Frequency Issue by Resampling**
    new_sfreq = max(100, 2 * max([high for _, high in eeg_bands.values()]))  # Ensure at least 90 Hz
    epochs.resample(new_sfreq, npad="auto")  # Resample to fix Nyquist limit

    # üìâ **Use a Subset of Epochs for Faster Computation**
    n_epochs = min(100, len(epochs))  # Use max 100 epochs
    epochs = epochs[:n_epochs]

    raw_data = epochs.get_data()  # Shape: (n_epochs, n_channels, n_times)

    # ‚úÖ **Reduce Dimensionality: Compute Mean Across Channels**
    raw_data = np.mean(raw_data, axis=1)  # Shape: (n_epochs, n_times)

    # **Parallel Function to Compute Entropy**
    def compute_entropy(epoch):
        return {
            "ApEn": ant.app_entropy(epoch),
            "SampEn": ant.sample_entropy(epoch)
        }

    # **Compute Entropy for Each Band**
    for band, (low, high) in eeg_bands.items():
        filtered_data = mne.filter.filter_data(
            raw_data, sfreq=epochs.info['sfreq'], l_freq=low, h_freq=high, verbose=False
        )

        # üî• **Parallel Entropy Computation**
        entropy_results = Parallel(n_jobs=-1)(
            delayed(compute_entropy)(epoch) for epoch in filtered_data
        )

        # Convert list of dicts to separate lists
        ApEn_vals = [res["ApEn"] for res in entropy_results]
        SampEn_vals = [res["SampEn"] for res in entropy_results]

        # Store mean entropy values
        entropy_data.append({
            "Filename": file.split("/")[-1],
            "Band": band,
            "ApEn": np.mean(ApEn_vals),
            "SampEn": np.mean(SampEn_vals)
        })

# üìë **Convert to DataFrame**
df_entropy = pd.DataFrame(entropy_data)

# üîÑ **Convert from Wide to Long Format for Seaborn**
df_entropy_melted = df_entropy.melt(
    id_vars=["Filename", "Band"],
    var_name="Entropy Type",
    value_name="Entropy Value"
)

# üé® **Plot with Improved Readability**
plt.figure(figsize=(12, 6))

sns.barplot(
    x="Entropy Type",
    y="Entropy Value",
    hue="Band",
    data=df_entropy_melted,
    capsize=0.1,
    errcolor="black",
    errwidth=1.5
)

# üñå **Styling for Better Readability**
plt.title("Distribution of Approximate Entropy and Sample Entropy Across Healthy EEG Files", fontsize=14, fontweight='bold')
plt.xlabel("Entropy Type", fontsize=12)
plt.ylabel("Entropy Value", fontsize=12)
plt.xticks(rotation=15, fontsize=11)
plt.yticks(fontsize=11)
plt.legend(title="EEG Bands", fontsize=11, title_fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.6)

# üìà **Show the Plot**
plt.show()

# üìÅ **Save Data to Google Drive**
save_path = "/content/drive/MyDrive/ADSP_VS/Healthy_Entropy_Results.csv"
df_entropy.to_csv(save_path, index=False)
print(f"‚úÖ Entropy results saved to: {save_path}")
```

```python
import mne
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.signal import hilbert  # Hilbert Transform for PLV
from google.colab import drive  # Google Drive support

# üìÇ **Mount Google Drive**
drive.mount('/content/drive')

# üìÇ **Specify Healthy EEG .fif Files**
healthy_file_paths = [
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_1-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_2-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_3-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_4-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_5-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_6-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_7-epo-normalized.fif",
    "/content/drive/MyDrive/ADSP_VS/Healthy_Subject_8-epo-normalized.fif"
]

# üß† **Define EEG Bands (Hz)**
eeg_bands = {
    "delta": (1, 4),
    "theta": (4, 8),
    "alpha": (8, 12),
    "beta": (12, 30),
    "gamma": (30, 45)
}

# üìä **Initialize List for PLV Storage**
plv_data = []

# üöÄ **Process Each Healthy Subject File**
for file in healthy_file_paths:
    print(f"üîÑ Loading file: {file}")
    epochs = mne.read_epochs(file, preload=True)

    # üîΩ **Fix Nyquist Frequency Issue by Resampling**
    new_sfreq = max(100, 2 * max([high for _, high in eeg_bands.values()]))
    epochs.resample(new_sfreq, npad="auto")

    # üìâ **Use a Subset of Epochs for Faster Computation**
    n_epochs = min(100, len(epochs))
    epochs = epochs[:n_epochs]

    raw_data = epochs.get_data()  # Shape: (n_epochs, n_channels, n_times)

    # üß† **Compute PLV for Functional Connectivity**
    def compute_plv(data, sfreq, low, high):
        """Compute PLV between all channel pairs for a given frequency band."""
        filtered = mne.filter.filter_data(data, sfreq=sfreq, l_freq=low, h_freq=high, verbose=False)
        analytic_signal = hilbert(filtered)  # Apply Hilbert Transform
        phase_angles = np.angle(analytic_signal)  # Extract phase

        n_channels = phase_angles.shape[0]
        plv_matrix = np.zeros((n_channels, n_channels))  # PLV storage

        # Compute PLV for each channel pair
        for i in range(n_channels):
            for j in range(i + 1, n_channels):
                phase_diff = phase_angles[i] - phase_angles[j]
                plv = np.abs(np.mean(np.exp(1j * phase_diff)))  # Compute PLV
                plv_matrix[i, j] = plv
                plv_matrix[j, i] = plv  # Symmetric matrix

        return np.mean(plv_matrix[np.triu_indices(n_channels, k=1)])  # Mean PLV

    # **Compute PLV for Each EEG Band**
    for band, (low, high) in eeg_bands.items():
        plv_value = compute_plv(raw_data, epochs.info['sfreq'], low, high)

        plv_data.append({
            "Filename": file.split("/")[-1],
            "Band": band,
            "PLV": plv_value
        })

# üìë **Convert to DataFrame**
df_plv = pd.DataFrame(plv_data)

# üé® **Plot PLV Values**
plt.figure(figsize=(12, 6))
sns.barplot(
    x="Band",
    y="PLV",
    data=df_plv,
    capsize=0.1,
    errcolor="black",
    errwidth=1.5
)
plt.title("Phase Locking Value (PLV) Across EEG Bands for Healthy Subjects", fontsize=14, fontweight='bold')
plt.xlabel("EEG Band", fontsize=12)
plt.ylabel("Mean PLV", fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.show()

# üìÅ **Save Data to Google Drive**
save_path = "/content/drive/MyDrive/ADSP_VS/Healthy_PLV_Results.csv"
df_plv.to_csv(save_path, index=False)
print(f"‚úÖ PLV results saved to: {save_path}")
```
